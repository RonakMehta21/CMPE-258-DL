* Novelty Detection: It detects any new instance that is very different from the training set. This algorithm should be feeded with very "clean" training dataset. For example, if the training dataset has thousands of dogs and 1% of them are chihuahuas. Then the algorithm should detect these chihuahuas.

* Most semi-supervised algorithms are combinations of supervised and unsupervised algorithms. For example, deep belief networks (DBNs) are based on unsupervised components called restricted boltzmann machines(RBMs). RBMs are sequentially trained in unsupervised manner and then fine-tuned using supervised learning techniques.

* Reinforcement Learning: Agent observes the environment and performs some action in a given situation. With every action the agent gets some rewards(or penalty for negative rewards). The agent then updates the strategy also called policy to get the optimal rewards.

* Batch Learning: In batch learning, the system is incapable of learning incrementally. The system has to be trained on the complete data offline while requiring lot of computational resources. After training, the system is launched into production with no further learning.

* Online learning: In online learning, the system is trained incrementally by feeing the data sequentially either individually or in small groups called mini-batches. Good for systems that receieve data in continuous flow.

* Online learning algorithms can also be used to train on huge dataset that cannot fir into machine's main memory called out-of-core learning. The algorithm loads part of the data and train on it and repeats until it has trained on the complete data. But usually the out-of-core learning is done offline.

## Line Magics
Line magics are functions that you can run on cells and take as an argument the rest of the line from where they are called. You call them by placing a '%' sign before the command. The most useful ones are:


`%matplotlib inline`: This command ensures that all matplotlib plots will be plotted in the output cell within the notebook and will be kept in the notebook when saved.

`%timeit`: Runs a line a ten thousand times and displays the average time it took to run it.

`%reload_ext autoreload`: Reload all modules before executing a new line. If a module is edited, it is not necessary to rerun the import commands, the modules will be reloaded automatically.

```
%timeit [i+1 for i in range(1000)]
```
